As, HDPK Dharmasena and S Ravi are the umpires in most of the matches, so filling null values with them.

```python
df.city.fillna("Dubai", inplace = True) # As matches have been played in Dubai International Cricket Stadium, So consider the city to be Dubai
df.fillna(0, inplace = True)
df.winner.fillna('Raining', inplace = True)
df.player_of_match.fillna('Raining', inplace = True)
df.umpire1.fillna('HDPK Dharmasena', inplace = True)
df.umpire2.fillna('S Ravi', inplace = True)
df.drop(['umpire3'], axis = 1, inplace = True)
```

```python
df.isnull().sum()
```

```python
print(df.season.unique(),
      "\nTotal number of IPL seasons:", df.season.nunique(),
      "\nTeams:\n", np.unique(df[['team1','team2']].values))
```

#### Number of matches played each season

```python
def annot_plot(ax,w,h): # function to add data to plot
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    for p in ax.patches:
        ax.annotate(f'{p.get_height()}' ,(p.get_x() +w, p.get_height()+h))
    
    
```

```python
# plt.figure(figsize=(10,5))
# plt.title("Number of matches played each season")
# ax = sns.countplot(x= 'season', data = df)
# annot_plot(ax, 0.15,0.5 )

# plt.show()

plt.figure(figsize=(12,8))
plt.title('Number of Matches played in each IPL seasons')
ax=sns.countplot(x=matches.season)
annot_plot(ax,0.08,1)
plt.show()
```

From the above graph, we can see that 76(max) matches took place in the year 2013 

```python
#Let's Check the venue where the most of the IPL matches has taken part
df.venue.value_counts()
```

```python
# df.venue.value_counts().sort_values(ascending = True).plot(kind = 'barh', figsize=(15,15), fontsize=15, color='red')
plt.figure(figsize=(12,6))
sns.countplot(x='venue', data=df)
plt.xticks(rotation='vertical')
plt.show()
```

From the above graph it's very clear that most of the IPL took place in the Eden Gardens

```python
df['winner'].value_counts().sort_values(ascending =False).plot(kind = 'barh', figsize=(10,5), fontsize=15, color='red')
```

We can clearly see that Mumbai Indians won the maximum IPL.

```python
# Matches won w.r.t to the seasons
plt.figure(figsize=(12,6))
sns.countplot(x='season', hue='winner', data=df)
plt.xticks(rotation='vertical')
plt.show()
```

The graph above predicts in which season which one of the teams has won the match

```python
## Toss Decision
toss_dec = df.toss_decision.value_counts().plot(kind = 'bar', color='pink')
toss_dec
```

We can see that whenever a team wins the toss choose fielding. 

```python
# let's see the chances of winning after the winning the toss 
df['team_toss_win'] = np.where((df.toss_winner == df.winner), 1,0)
plt.figure(figsize=(10, 5))
sns.countplot('team_toss_win', data = df, hue = 'toss_decision')
plt.title("Toss Wins vs Match Wins")
plt.xlabel("Winning Toss and winning match")
plt.ylabel("Frequency")
```

```python
## let's see the top 5 man of the match
player_of_match=df['player_of_match'].value_counts().head(5).plot(kind = 'bar', color = 'g')
player_of_match
```

```python
ax= df.groupby('season')['batsman_runs'].sum().sort_values(ascending= False).plot(kind = 'bar', color = "orange")
annot_plot(ax,0.08,1)
plt.show()
```

Teams scored the msot in 2013 and least in 2009.

```python
## Top 5 run scores
ax= df.groupby('batsman')['batsman_runs'].agg("sum").sort_values(ascending= False).head(5).plot(kind = 'bar', color = "purple")
annot_plot(ax,0.08,1)
plt.show()
```

The above graph shows the top run scores in IPL, Virat Kohli is leading the chart followed by SK Raina.

```python
ax= df.groupby('bowler')['ball'].agg("sum").sort_values(ascending= False).head(5).plot(kind = 'bar', color = "purple")
annot_plot(ax,0.08,1)
plt.show()
```

Harbhajan Singh is the the bowler with most number of balls bowled in IPL matches.

```python
ax= df.groupby('bowler')['extra_runs'].agg("sum").sort_values(ascending= False).head(5).plot(kind = 'bar', color = "purple")
annot_plot(ax,0.08,1)
plt.show()
```

Malinga tops the chart with 293 extra runs followed by Pravin Kumar.

```python
ax= df.groupby('bowler')['player_dismissed'].count().sort_values(ascending= False).head(10).plot(kind = 'bar', color = 'brown')
annot_plot(ax,0.08,1)
plt.title("Top Wicket Takers of IPL")
plt.xlabel("Bowler")
plt.ylabel("Total Wickets Taken")
plt.show()
```

Harhajan Singh has taken the most number of wickets in IPL followed by Mishra and Chawla
